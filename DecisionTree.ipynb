```python
def calculateEntropy(a,b) :
    #print("A",a,"B",b)
    attributeList = []
    if a == b :
        attributeList.append(1.0)
    elif a == 0 or b == 0:
        attributeList.append(0)
    else:
        aBYsum = a/(a+b)
        bBYsum = b/(a+b)
   
        output = -(aBYsum *  np.log2(aBYsum)) - (bBYsum * np.log2(bBYsum))
        attributeList.append(output)
    attributeList.append(a)
    attributeList.append(b)
    
    return attributeList
#calculateEntropy(4,0)
```

```python
def calculateVarianceImpurity(a,b) :
    attributeList = []
    if a == b :
        attributeList.append(0.25)
    elif a == 0 or b == 0:
        attributeList.append(0)
    else:

        output = (a * b)/(a+b)**2
        attributeList.append(output)
    attributeList.append(a)
    attributeList.append(b)

    return attributeList
print(calculateVarianceImpurity(5,5))
```

```python
class DecisionTree():
      
    def __init__(self,key,level):
        self.left = None
        self.right = None
        self.key = key
        self.level = level
        self.leftValue = None
        self.rightValue = None
        self.value = None

    def getLeftChild(self):
        return self.left
    def getRightChild(self):
        return self.right
    
    def setNodeValue(self,value):
        self.value = value

    def setLeftValue(self, value):
        self.leftValue = value

    def getLeftValue(self):
        return self.leftValue

    def getRightValue(self):
        return self.rightValue

    def setRightValue(self, value):
        self.rightValue = value

        
    def getNodeValue(self):
        if self.key != None:
            return self.key
        else:
            return None
    
    def getNodeLevel(self):
        return self.level
    
    def getNodeBranch(self):
        return self.branch

    def insertRight(self,newNode):
        if self.right == None:
            self.right = newNode
       

    def insertLeft(self,newNode):
        if self.left == None:
            self.left = newNode
# test tree

```

```python
## out is a list containing 4 values first two indices have +ve and -ve examples for branch 1
## next two indices have same for the zero branch
## we calculate the information gain for this particular attribute, pAndN has total number of examples or total class count
def calculateAttributeEntropy(out,pAndN,classEntropy):
    #print("Out Values:",out)
    #print("TotalCount:",pAndN)
    gainForOne = []
    gainForZero = []
    totalGain = 0
    #if ((out[0] > 0 or out[1] > 0) and  (out[2] > 0 or out[3] > 0)):
    #print("For Above attribute OUT[0]",out[0],"out[1]",out[1])
    if (heuristic == 'IG'):
        gainForOne = calculateEntropy(out[0],out[1])
        gainForZero = calculateEntropy(out[2],out[3])
    else:
        gainForOne = calculateVarianceImpurity(out[0],out[1])
        gainForZero = calculateVarianceImpurity(out[2],out[3])
    #print("For Above attribute OUT[2]",out[2],"out[3]",out[3])
     
    #print("GZERO:",len(gainForZero),"GONE:",len(gainForOne))
    if (len(gainForZero) > 0 and len(gainForOne) > 0):
        totalGain = (((out[0] + out [1])/pAndN)*(gainForOne[0])) + (((out[2] + out [3])/pAndN) * gainForZero[0])
    #print("G1:", gainForOne[0])
    #print("G0:", gainForZero[0])
    #print("TotalGain", totalGain)
    #print("ClassEntropy", classEntropy)
    #print(totalGain)
    #print(type(classEntropy))
    #print(type(totalGain))
    entropyForAttribute = classEntropy - totalGain
    return entropyForAttribute
```
